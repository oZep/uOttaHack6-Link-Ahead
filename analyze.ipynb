```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import statistics
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv('./dataset_cybersecurity_michelle.csv')
for i in df.keys():
    print(i)
```

```python
for column_name in df.keys():
    if column_name.startswith("qty") and not "servers" in column_name and not "resolved" in column_name and not "redirect" in column_name:
        if "url" in column_name:
            divisor = "length_url"
        else:
            divisor = column_name.split("_")[-1] + "_length"
        df[column_name] /= df[divisor]
df
```

```python
df_new = df[df['phishing'] == 1]
```

```python
def createHist(key):
    df[key].hist( bins=25, density=True, alpha=0.6, color='b')
    plt.title('Distribution of ' + key)
```

```python
k = df_new.keys()
show = k[5]
createHist(show)   
```

```python
means_and_variances = []
for column_name in df_new:
    means_and_variances.append(
        [column_name, df_new.mean(), df_new[column_name].std()]
    )
```

```python
with open('dict2.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for i in means_and_variances:
        writer.writerow(i)
```
